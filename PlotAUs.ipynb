```python
import dlibPreprocess
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
modelFolder = "/Users/kristshingjergji/Documents/PhD/Experiment_1/models"
predictor = dlibPreprocess.openFiles(modelFolder)[0]
image = cv2.imread('')
```

```python
plt.imshow(image)
```

```python
landmarks, rects = dlibPreprocess.faceLandmarks(image, predictor)
```

```python
rotatedImage, rotatedLandmarks = dlibPreprocess.alignFromLandmarks(image, predictor)
```

```python
resizedCroppedImage, resizedCroppedLandmarks = dlibPreprocess.cropResize(image, 112, predictor)
```

```python
feature_cbd = dlibPreprocess.hogsLandmarks(resizedCroppedImage, resizedCroppedLandmarks, modelFolder)
```

```python
AUs = dlibPreprocess.AUdetection(feature_cbd, modelFolder)
```

```python
dlibPreprocess.plotAUs(image, AUs, landmarks,rects)
```
